```python
from rdkit import Chem
from rdkit.Chem import Draw

# Define the SMILES string
# Example: PAO-OH
smiles = 'C1(O)=C(C(C)(C)(C))C=CC=C1C(C)(C)(C)'

# Convert SMILES to a molecule object
mol = Chem.MolFromSmiles(smiles)

# Draw the molecule
Draw.MolToImage(mol)
```
