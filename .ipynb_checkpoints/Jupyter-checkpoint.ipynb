```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
from IPython.display import Image

# Define the SMILES string
smiles = 'CC(C)(C)C1=CC(\C=C/CO)=CC(=C1O)C1=CC(\C=C/CO)=CC(=C1O)C(C)(C)C'

# Convert SMILES to a molecule object
mol = Chem.MolFromSmiles(smiles)

# Generate 2D coordinates for the molecule
AllChem.Compute2DCoords(mol)

# Set up drawing options for a better layout
drawer = Draw.MolDraw2DCairo(300, 300)  # Adjust dimensions as needed
options = drawer.drawOptions()
options.fixedBondLength = 30  # Adjust for a more compact drawing
options.useBWAtomPalette()  # Black-and-white palette, if desired

# Draw the molecule and save to a file
drawer.DrawMolecule(mol)
drawer.FinishDrawing()
dirr=r'C:\Users\arup2\OneDrive - University of California Merced\Desktop\LAMMPS'
drawer.WriteDrawingText(dirr+"\\molecule_image.png")  # Save to a file

# Display the image
Image(dirr+"\\molecule_image.png")

```
